```python
date = df_temiz["InvoiceDate"].max() + dt.timedelta(days=1) 
```

```python
# Receny, Frequency, Monetary

rfm = df_temiz.groupby("Customer_ID").agg({
    "InvoiceDate": lambda x: (date - x.max()).days,   # Recency
    "Invoice": "nunique",                             # Frequency
    "TotalRevenue": "sum"                             # Monetary
}).reset_index()
```

```python
rfm["R_Score"] = pd.qcut(rfm["Recency"], 5, labels=[5, 4, 3, 2, 1])
rfm["F_Score"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5]) 
rfm["M_Score"] = pd.qcut(rfm["Monetary"],5, labels=[1, 2, 3, 4, 5])
# Bin edges must be unique(Frequency alanÄ±nda aynÄ± deÄŸere sahip Ã§ok kayÄ±t olduÄŸundan Ã¶tÃ¼rÃ¼ qcut iÅŸlemi uygulanmÄ±yordu.) hatasÄ±nÄ± engellemek amacÄ± ile Frequency deÄŸiÅŸkenlerine rank(method="first") ekledik.
```

```python
# Elbow methodu

ssd = []
K = range(2, 12)

for k in K:
    kmeans = KMeans(n_clusters=k).fit(rfm_segment)
    print(kmeans.inertia_)
    ssd.append(kmeans.inertia_)
    
    
plt.plot(K, ssd, "bx-")
plt.xlabel("Residuals Distance")
plt.title("Optimum Number of Clusters")
plt.show()    
```
